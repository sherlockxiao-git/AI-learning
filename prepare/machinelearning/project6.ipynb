```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from PIL import Image
```

```python
data = pd.read_csv("ex7data2.txt",header=None,names=["X1","X2"])
data.head()
```

```python
fig,ax = plt.subplots(figsize=(8,8))
ax.scatter(data["X1"],data["X2"],s=50,c="b",marker="o")
ax.set_xlabel("X1")
ax.set_ylabel("X2")
plt.show()
```

```python
X=data[["X1","X2"]].values.reshape(300,2)
```

```python
km=KMeans(n_clusters=3,n_init=15)
```

```python
km2=KMeans(n_clusters=2,n_init=15)
km4=KMeans(n_clusters=4,n_init=15)
km5=KMeans(n_clusters=5,n_init=15)
```

```python
km.fit(X)
```

```python
km2.fit(X)
km4.fit(X)
km5.fit(X)
```

```python
c=km.labels_
```

```python
c2=km2.labels_
data["C2"]=c2
c4=km4.labels_
data["C4"]=c4
c5=km5.labels_
data["C5"]=c5
```

```python
data["C"]=c
```

```python
data
```
