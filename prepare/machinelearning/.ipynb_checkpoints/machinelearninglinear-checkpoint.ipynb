```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("ex1data1.txt",names=["Population","Profit"])
```

```python
data.describe()
```

```python
data.plot(kind="scatter",x="Population",y="Profit",figsize=(12,8))
plt.show()
```

```python
data.insert(0,"Ones",1)
X=np.array(data.loc[:,["Ones","Population"]])
```

```python
Y=np.array(data.loc[:,["Profit"]])
print(Y.shape)
```

```python
# Y=np.array(data["Profit"])
# Y.shape
```

```python
theta=np.full((1,2),0.1)
```

```python
hx=np.dot(X,theta.T)
```

```python
err=hx-Y
```

```python
m=X.shape[0]
alpha=0.01
```

# 计算损失函数

```python
def J(theta,X,Y):
    return np.sum((Y-X.dot(theta))**2)/(X.shape[0]*2)
J(theta.T,X,Y)
```

# 更新theta参数值

## 使用函数封装

```python
def dJ2(theta, X_b, y,alpha):
    res = np.empty(X_b.shape[1])
    for i in range( len(theta)):
        res[i] = theta[i] - alpha*(np.sum((X_b.dot(theta) - y)*(X_b[:,i].reshape(-1,1)))/ X_b.shape[0])
    return res 
dJ2(theta.T,X,Y,alpha)
```

## 更新theta 使用numpy特性直接合并

```python
thetanew=theta
```

```python
thetanew=thetanew-np.sum(err*X,axis=0)/m*alpha
thetanew.shape
```

# 迭代函数

```python
def gradient_descent(X, Y, initial_theta, eta, n_iters = 1000, epsilon=1e-8):
    cost=np.zeros((n_iters,1))
    theta = initial_theta
    cur_iter = 0

    while cur_iter < n_iters:
        cost[cur_iter]=J(theta,X,Y)
        last_theta = theta
        theta=theta-(np.sum((np.dot(X,theta)-Y)*X,axis=0)/m*alpha).reshape(2,1)
        if(abs(J(theta, X, Y) - J(last_theta, X, Y)) < epsilon):
            break
            
        cur_iter += 1

    return cost,theta
```

```python
cost,thetanew=gradient_descent(X,Y,theta.T,alpha)
```

```python
fig,ax=plt.subplots(figsize=(12,8))
population=np.linspace(data["Population"].min(),data["Population"].max(),100)
profit=thetanew[0,0]+(thetanew[1,0]*population)
ax.plot(population,profit,"r",label="Prediction")
ax.scatter(data["Population"],data["Profit"],label="Traning Data")
ax.set_xlabel("Population")
ax.set_ylabel("Profit")
ax.set_title("Predicted Profit vs. Population Size")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(np.arange(1000), cost.reshape(1000), 'r')
ax.set_xlabel('Iterations')
ax.set_ylabel('Cost')
ax.set_title('Error vs. Training Epoch')
plt.show()
```

# 自行实现算法

# 改进的算法

```python
def J(theta,X,Y):
    return np.sum((Y-X.dot(theta))**2)/(X.shape[0]*2)
```

```python
def dJ(theta, X_b, y):
    res = np.empty(len(theta))
    res[0] = np.sum(X_b.dot(theta)-y )
    for i in range(1, len(theta)):
        res[i] = np.sum((X_b.dot(theta) - y)*(X_b[:,i].reshape(-1,1)))

    return res / len(X_b)
dJ(theta.T,X,Y)
```

```python
def gradient_descent3(X_b, y, initial_theta, eta, n_iters = 1e4, epsilon=1e-8):
    
    theta = initial_theta
    cur_iter = 0
    
    while cur_iter < n_iters:
        gradient = dJ(theta, X_b, y).reshape(2,1)
        last_theta = theta
        theta = theta - eta * gradient
        if(abs(J(theta, X_b, y) - J(last_theta, X_b, y)) < epsilon):
            break
            
        cur_iter += 1

    return theta
```

```python
t3=gradient_descent3(X,Y,theta.T,alpha)
t3
```

```python
fig,ax=plt.subplots(figsize=(12,8))
population=np.linspace(data["Population"].min(),data["Population"].max(),100)
profit=t3[0,0]+(t3[1,0]*population)
ax.plot(population,profit,"r",label="Prediction")
ax.scatter(data["Population"],data["Profit"],label="Traning Data")
ax.set_xlabel("Population")
ax.set_ylabel("Profit")
ax.set_title("Predicted Profit vs. Population Size")
plt.show()
```
